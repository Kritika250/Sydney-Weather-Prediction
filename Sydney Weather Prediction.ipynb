### 3. What were the available ML model options you had to perform this task?
The available machine learning models I considered for this task included:

* Logistic Regression: A baseline classification algorithm that models the probability of a binary outcome.
* Linear Discriminant Analysis (LDA): A linear classification technique that finds a linear combination of features to separate classes.
* K-Nearest Neighbors (KNN): A non-parametric algorithm that classifies based on the majority vote of the nearest neighbors, with Grid Search used to optimize hyperparameters.
* Decision Tree: A tree-based method that splits the data based on feature values, creating a model that predicts the target variable.
* Bagging: An ensemble method that creates multiple versions of a classifier and aggregates them to improve model performance and reduce variance.
* Random Forest: An extension of decision trees that uses multiple trees to create a stronger, more accurate model.
* Gradient Boosting: An ensemble technique that sequentially builds models by correcting the errors of the previous model.
* AdaBoost: Another boosting algorithm that adjusts the weights of misclassified data points to improve accuracy, including a variation using Random Forest as the base estimator.
* XGBoost: A high-performance gradient boosting algorithm that focuses on both speed and accuracy, commonly used in competitive machine learning tasks.
These models were selected to provide a mix of linear, non-linear, and ensemble-based methods, allowing me to evaluate and compare their effectiveness in predicting rainfall.

### Missing Values Imputation

```python
# Columns to be imputed with mode
columns_to_impute = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                     'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm',
                     'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RainToday']

# Impute missing values with mode for the specified columns
df[columns_to_impute] = df[columns_to_impute].apply(
    lambda x: x.fillna(x.mode()[0]) if x.isna().sum() > 0 else x
)
```

### Outliers Treatment

```python
#Creating box plots to check for outliers
numerical_columns = ['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
                     'Humidity9am', 'Humidity3pm', 'Pressure9am', 'Pressure3pm',
                     'Cloud9am', 'Cloud3pm', 'Temp9am', 'Temp3pm', 'RainToday','RainTomorrow']

# Set the figure size and adjust space
plt.figure(figsize=(16, 12))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

# Loop through numerical columns and create a 4x4 grid of box plots
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(4, 4, i)  # Adjust the grid to 4x4 to fit 13 columns
    sns.boxplot(data=df, y=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Outliers are in MaxTemp,Rainfall,Evaporation,Humdity9am,Humdity3pm,Pressure9am,Pressure3pm,Temp9am,Temp3pm)
```

```python
# Function for outlier treatment using IQR method
def treat_outliers_iqr(column):
    Q1 = column.quantile(0.25)  # First quartile (25th percentile)
    Q3 = column.quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound for outliers
    upper_bound = Q3 + 1.5 * IQR  # Upper bound for outliers
    
    # Capping outliers
    return column.apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

# Columns with identified outliers
outlier_columns = ['MaxTemp', 'Rainfall', 'Evaporation', 'Humidity9am', 'Humidity3pm',
                   'Pressure9am', 'Pressure3pm', 'Temp9am', 'Temp3pm']

# Apply the IQR method to treat outliers only in these specific columns
df[outlier_columns] = df[outlier_columns].apply(treat_outliers_iqr)

# Verify if the outliers have been treated
df[outlier_columns].describe()
```
